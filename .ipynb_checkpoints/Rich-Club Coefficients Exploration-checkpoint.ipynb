```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import json
import scipy.stats as stats

%matplotlib inline
```

```python
with open('results/rcc_normalized.json') as json_file:
    results = json.load(json_file)
```

```python
projects = pd.read_csv('project_graph_dimension.csv')
projects.head()
```

```python
missing = []
max_values = []
for pid in projects['id']:
    project_dict = results[str(pid)]
    if project_dict == {}:
        missing.append(pid)
    else:
        k_list = project_dict['k']
        rcc_list = project_dict['phi']
        
        max_rcc = max(rcc_list)
        k = k_list[rcc_list.index(max_rcc)]
        
        max_values.append(tuple((pid, k, max_rcc)))
```

```python
all_rcc = []
for pid in projects['id']:
    project_dict = results[str(pid)]
    if project_dict == {}:
        missing.append(pid)
    else:
        k_list = project_dict['k']
        rcc_list = project_dict['phi']
        
        all_rcc = all_rcc + rcc_list
```

```python
plt.figure(figsize=(12, 8))
plt.hist(all_rcc, bins=20)
plt.xlabel('$\\rho$', fontsize=13)
plt.ylabel('Counts', fontsize=13)
plt.grid();
```

```python
max_list = {}
for index, row in projects[['id', 'name']].iterrows():
    pid = row['id']
    name = row['name']
    project_dict = results[str(pid)]
    if project_dict == {}:
        missing.append(pid)
    else:
        k_list = project_dict['k']
        rcc_list = project_dict['phi']
        
        max_coeff = max(rcc_list)
        max_list[pid] = max_coeff
```

```python
df = pd.DataFrame.from_dict(max_list, orient='index')
df.columns = ['Maximum Rich-club coefficient']
```

```python
df.sort_values(by='Maximum Rich-club coefficient').plot.bar(figsize=(16, 8), rot=90)
plt.axhline(y=1, color='black', linestyle='--', label='random network')
plt.grid()
plt.legend();
```

### Update projects file with max rcc and correspondent k

```python
out_df = projects.merge(pd.DataFrame(max_values, columns=['id', 'k', 'rcc']), on='id', how='outer')
out_df.sort_values(by='rcc', ascending=False)
```

```python
out_df.to_csv('results/projects_max_rcc.csv', index=None)
```

### Normalize each project distribution of $\rho$ independently

```python
max_values_z = []
for pid in results:
    if results[pid] != {}:
        proj_distr = results[pid]['phi']
        mu = np.mean(proj_distr)
        sigma = np.std(proj_distr)
        z = [(x - mu)/sigma for x in proj_distr]
        
        max_rcc = max(z)
        k = k_list[z.index(max_rcc)]
        
        max_values_z.append(tuple((int(pid), k, max_rcc)))
```

```python
out_df_z = projects.merge(pd.DataFrame(max_values_z, columns=['id', 'k', 'z-score']), on='id', how='outer')
out_df_z.sort_values(by='z-score', ascending=False)
```
